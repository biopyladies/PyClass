# **Funções em Python** 🐍


## **1. O que são funções em Python?** 🧐  

>**As funções são blocos de código reutilizáveis que ajudam a tornar os programas mais organizados e eficientes.**

**Elas servem para:**  

  1. **Estruturar o código em partes menores e compreensíveis** – Assim como dividimos um experimento em etapas, um programa pode ser dividido em funções para facilitar o entendimento e a manutenção.  

  2. **Evitar repetições** – Em vez de escrever o mesmo código várias vezes, podemos criar uma função e simplesmente "chamá-la" quando necessário.  

  3. **Tornar o código mais legível e reutilizável** – Um código bem organizado, com funções bem definidas, é mais fácil de entender, modificar e compartilhar com outras pessoas.  

>💡 **Exemplo na prática:** Pense em uma função como uma receita/protocolo de laboratório. Uma vez definida, você pode reutilizá-la várias vezes sem precisar reescrever todo o procedimento!

### **A função `len()`** ⏳  

>A função **`len()`** é utilizada para determinar a quantidade de elementos em um objeto. Em outras palavras, ela nos informa o tamanho de:  

1. **Strings** – Descobre quantos caracteres uma palavra ou frase possui.  
2. **Listas, tuplas e dicionários** – Conta quantos itens estão armazenados nessas estruturas.  



### **Como utilizar uma função em Python?** 🖥️  

- Para chamar (ou executar) uma função, seja ela **embutida no Python ou criada por você**, basta seguir estes passos:  

  1. **Escrever o nome da função** corretamente.  
  2. **Abrir e fechar parênteses** logo após o nome (sem espaço entre eles).  
  3. **Passar os argumentos** necessários dentro dos parênteses (se houver).  

💡 **Sintaxe geral:**  

```python

```

### **Exemplo 1** 🙂  

>Criamos uma função chamada **`verificar_proteina`**, que recebe como parâmetro uma sequência de aminoácidos, representada pela variável **`sequencia_aminos`**.  

*A função verifica dois critérios essenciais para validar a sequência proteica:*

1. **Comprimento da sequência** – Deve conter **pelo menos 10 aminoácidos** 🧮.  
2. **Início correto da sequência** – O primeiro aminoácido deve ser **Metionina (M)**, pois na maioria dos seres vivos 🦕 as proteínas iniciam com esse aminoácido.  

💡 **Por que Metionina?**  

A tradução de proteínas começa com o códon **AUG**, que codifica a **Metionina (M)**.

>Para saber mais sobre esse processo, confira este artigo:  
📚 [Etapas da tradução – Khan Academy](https://pt.khanacademy.org/science/biology/gene-expression-central-dogma/translation-polypeptides/a/the-stages-of-translation)  

````python
# @title **Implementação da função:**
def verificar_proteina(sequencia_aminos: str) -> None:
    """Verifica se uma sequência de aminoácidos é uma proteína válida."""
    if len(sequencia_aminos) < 10 or sequencia_aminos[0] != "M":
        print("Essa proteína não é válida.")
    else:
        print("A proteína é válida!")
````

>Para executar esta função, precisamos seguir as mesmas regras de executar funções embutidas: **nome da função, parênteses e parâmetros**

````python
# @title **Exemplo de uso:**
verificar_proteina("MLTAEDKKLIQQAWEKAASHQEEFGAEALTRMFTTYPQTKTYFPHFDLSPGSDQVRGHGKKVLGALGNAVKNVDNLSQAMAELSNLHAYNLRVDPVNFKLLSQCIQVVLAVHMGKDYTPEVHAAFDKFLSAVSAVLAEKYR")
````

A sequência de aminoácidos utilizada para testarmos nossa função é a **sequência da subunidade A da hemoglobina de galos** 🐔.

**Para verificar mais sobre essa sequência**: [link](https://www.ncbi.nlm.nih.gov/protein/1HBR_A)

Crie uma função que:  
- Se chame **`verificar_complementariedade`**  
- Receba **dois parâmetros**: `nucleotideo1` e `nucleotideo2`  

A função deve verificar se os nucleotídeos fornecidos são **complementares** de acordo com a **complementariedade de bases do DNA** 🧬:  

🔹 **Adenina (A) se liga à Timina (T)**  
🔹 **Citosina (C) se liga à Guanina (G)**  

Se os nucleotídeos forem complementares, a função deve retornar **`True`**, caso contrário, deve retornar **`False`**.  

>📚 Para saber mais sobre esse conceito fundamental da biologia molecular, confira este artigo:  
🔗 [Código Genético – Mundo Educação](https://mundoeducacao.uol.com.br/biologia/codigo-genetico.htm)  

💡 **Exemplo esperado:**  
```python
print(verificar_complementariedade("A", "T"))  # True
print(verificar_complementariedade("C", "G"))  # True
print(verificar_complementariedade("A", "G"))  # False
```

````python
# Responda aqui

# def
````

````python
# @title **Exemplo de implementação**
def verificar_complementariedade(nucleotideo1: str, nucleotideo2: str) -> bool:
    """Verifica se dois nucleotídeos são complementares no DNA."""
    pares_complementares = {"A": "T", "T": "A", "C": "G", "G": "C"}

    return pares_complementares.get(nucleotideo1) == nucleotideo2
````

### **Exemplo 2** 🙂  

>Criamos uma função chamada `is_nucleotide_valid`, que recebe um **único nucleotídeo** como entrada e verifica se ele é válido no contexto do **DNA**.  

1. Os nucleotídeos válidos são: **A (Adenina), C (Citosina), T (Timina) e G (Guanina).**  
2. O código converte a entrada para maiúscula para evitar erros de diferenciação entre **"a" e "A"**.    

````python
# @title **Implementação da função**
def is_nucleotide_valid(nucleotide: str) -> None:
    """
    Verifica se um nucleotídeo é válido no DNA.

    Parâmetros:
    nucleotide (str): O nucleotídeo a ser verificado.

    Retorna:
    None (exibe uma mensagem informando se o nucleotídeo é válido ou não).
    """
    valid_nucleotides = {"A", "C", "T", "G"}
    nucleotide = nucleotide.upper()

    if nucleotide in valid_nucleotides:
        print(f"{nucleotide} é um nucleotídeo válido.")
    else:
        print(f"{nucleotide} NÃO é um nucleotídeo válido.")
````

## **Return**

- **A instrução/comando `return` finaliza a execução e retorna um valor da função.**

- **É recomendado utilizar `return` ao invés de `print()` dentro de funções em códigos maiores.**

  - **`return`**: Retorna um valor da função, permitindo que o valor seja usado em outras partes do programa. **Isso torna a função mais flexível e reutilizável.**
  - **`print()`**: Exibe uma saída no console, mas não retorna um valor. **É útil para depuração ou para scripts simples, mas limita a reutilização da função em contextos diferentes.**

### **Uso do `return` com valores booleanos** ✅❌

Na construção de **scripts e pipelines**, é muito comum utilizar `return` para retornar **valores booleanos (`True` ou `False`)**. Isso permite que o código **tome decisões automaticamente**, como continuar ou interromper um processo com base na resposta de uma função.  

💡 **Exemplo prático:**  
>Imagine que estamos analisando um **arquivo FASTA** 🧬 (um formato muito usado em bioinformática para armazenar sequências de DNA ou proteínas). Queremos verificar se um **nucleotídeo específico** (A, C, T ou G) é válido antes de contar quantas vezes ele aparece no arquivo.  

Nesse caso, faz mais sentido que nossa função **retorne `True` ou `False`**, para que o script possa decidir se continua ou para.  

📚 Para saber mais sobre arquivos FASTA, confira:  
🔗 [O que é um arquivo FASTA? – Varsomics](https://www.reviversoft.com/pt/file-extensions/fasta)  




````python
# @title **Implementação da função**
def is_nucleotide_valid(nucleotide: str) -> bool:
    """Verifica se um nucleotídeo é válido e retorna True ou False."""
    valid_nucleotides = {"A", "C", "T", "G"}  # Conjunto com nucleotídeos válidos
    return nucleotide.upper() in valid_nucleotides  # Retorna True ou False
````

````python
# Uso da função dentro de um script
nuc = input("Digite um nucleotídeo para verificação: ")
if is_nucleotide_valid(nuc):
    print(f"{nuc} é válido. Continuando a análise do arquivo FASTA...")
else:
    print(f"{nuc} NÃO é um nucleotídeo válido. Encerrando o script.")
````

🔹 **Por que usar `return` em vez de `print`?**  
✔️ Permite **usar o resultado** da função em outras partes do código.  
✔️ **Facilita a automação** de scripts e pipelines.  
✔️ **Evita saídas desnecessárias** no terminal, tornando o código mais limpo.


### **Parâmetros padrão em funções** ⚙️  

>Em Python, podemos definir **parâmetros padrão** para uma função. Isso significa que, se um valor **não for fornecido** ao chamar a função, o Python usará automaticamente o valor **pré-definido**.  
---

💡 **Como definir um valor padrão?**  

>Basta atribuir um valor ao parâmetro usando '`=`' ao definir a função.

````python
# @title **Definindo 'A' como valor padrão para `nucleotide`**
def is_nucleotide_valid(nucleotide="A"):
    """Verifica se um nucleotídeo é válido. Caso nenhum seja informado, assume 'A'."""
    valid_nucleotides = {"A", "C", "T", "G"}
    return nucleotide.upper() in valid_nucleotides
````

````python
print(is_nucleotide_valid("G"))
````

````python
print(is_nucleotide_valid())      # usa o valor padrão "A"
````

````python
print(is_nucleotide_valid("X"))
````

✅ **Evita erros** ao chamar a função sem argumentos.  
✅ **Facilita o uso da função** sem exigir sempre um valor explícito.  
✅ **Torna o código mais flexível**, permitindo que o usuário forneça um valor quando necessário.

### 🌟 **Agora é sua vez!**  

Crie uma função chamada **`calcular_tm`** que:

1. **Receba uma sequência de DNA (primer) como parâmetro**  
2. **Calcule a temperatura de anelamento (Tm)** utilizando a seguinte fórmula aproximada para primers curtos < 20 nucleotídeos:

````python
from IPython.display import display, Math

# Exibe a fórmula no Colab
display(Math(r"Tm = 4(G+C) + 2(A+T)"))
````

Onde:  

- **G** = número de guaninas no primer  
- **C** = número de citosinas  
- **A** = número de adeninas  
- **T** = número de timinas  

✅ **Retorne a Tm calculada**

💡 **Exemplo esperado:**  
```python
print(calcular_tm("AGCTGATCGGCA"))  
# Saída esperada: A temperatura de anelamento do primer AGCTGATCGGCA é aproximadamente 34°C.
```

````python
def calcular_tm(primer):
    """
    Calcula a temperatura de anelamento (Tm) de um primer curto (< 20 nucleotídeos)
    usando a fórmula: Tm = 4(G + C) + 2(A + T)

    Parâmetros:
    primer (str): sequência de DNA contendo os nucleotídeos A, T, C, G.

    Retorna:
    int: valor da temperatura de anelamento (Tm)
    """
    # Padroniza a sequência para letras maiúsculas
    primer = primer.upper()

    # Conta a quantidade de cada base
    G = primer.count('G')
    C = primer.count('C')
    A = primer.count('A')
    T = primer.count('T')

    # Aplica a fórmula de Tm
    tm = 4 * (G + C) + 2 * (A + T)

    return tm
````

````python
print(calcular_tm('TACTACCACCAAGATCT'))
````

````python
primer_exemplo = 'ACCCGCTGAACTTAAGC'
tm_resultante = calcular_tm(primer_exemplo)

print(f"A temperatura de anelamento do primer {primer_exemplo} é aproximadamente {tm_resultante}°C.")
````
