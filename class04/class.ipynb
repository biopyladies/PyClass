## Loops

Fazer um loop significa repetir alguma ao at que determinada condio seja satisfeita.

Hoje falaremos sobre dois tipos de loops do Python: `for` e `while`.

## `for` loops 

O loop `for`  uma forma de controle de fluxo usada para executar repetidamente um grupo de instrues por um nmero finito de vezes (loop definido) ou enquanto uma condio for satisfeita. Esse processo tambm  chamado de **iterao**. **Iterar** significa "*fazer ou dizer novamente; repetir, reiterar*".

### Entendendo a sintaxe

````python
apes = ["Homo sapiens", "Pan troglodytes", "Gorilla gorilla"]

for ape in apes:
    print(ape + " is an ape.")
````

### Vamos examinar cada parte do loop:

`for x in y`: `y`  o nome da lista na qual vamos rodar o loop e `x`  o nome que estamos dando a cada elemento pelo qual passa o loop.

`x` simplesmente um nome de uma varivel, mas se comporta um pouquinho diferente das variveis que vimos at aqui. Nos exemplos anteriores, ns crivamos uma varivel, armazenvamos algo nela e ela no mudava, a menos que escolhssemos fazer uma mudana. Quando criamos uma varivel para ser usada em um loop, no definimos seu valor: o valor vai ser automaticamente definido como cada elemento da lista e ser diferente cada vez que rodar o loop.

A primeira linha do loop termina com um `:` e todas as linhas seguintes so identadas - chamamos isso de **bloco de cdigo**. Nesse caso, o bloco identado  o **corpo** do loop e as linhas dentro dele sero executadas uma vez para cada elemento da lista. Quando queremos falar do elemento atual, usamos o nome da varivel que definimos na primeira linha.

O corpo do loop pode ter quantas linhas forem necessrias e nele podemos usar todas as funes e mtodos que j aprendemos, com uma exceo: **no podemos alterar a lista dentro do corpo do loop**.

Vamos exemplificar com um corpo mais complexo:

 Imagine que temos uma lista chamada apes que contm vrios nomes de primatas, como **["gorila", "chimpanz", "orangotango", "gibo"]**:

````python
for ape in apes:
    name_length = len(ape)
    first_letter = ape[0]
    print('\n', ape, " is an ape.")
    print("\n Its name starts with " + first_letter,
          "\n Its name has " + str(name_length) + " letters")

# O corpo desse loop tem 4 statements
````

 **DESCRIO NARRATIVA DO CDIGO**

*   A linha `for ape in apes:`  como se dissssemos: "Para cada item nesta lista de primatas, vou cham-lo temporariamente de ape e realizar algumas aes com ele."

*   Dentro do loop, `name_length = len(ape)` conta quantas letras tem o nome atual do primata e guarda esse nmero na varivel `name_length`

*   `first_letter = ape[0]` pega apenas a primeira letra do nome. O `[0]` significa "d-me o primeiro caractere" (em programao, comeamos a contar do zero).

*   `print('\n', ape, " is an ape.")` mostra na tela uma linha em branco `(\n)` seguida do nome do primata e a frase explicando que ele  um macaco.

*   O ltimo `print(`) mostra duas informaes:
Em uma nova linha: `Its name starts with [primeira letra]`

*   Em outra nova linha: `Its name has [nmero] letters`

*   O `str(name_length)` converte o nmero em texto para poder ser exibido.

Aps mostrar tudo para o primeiro primata, o programa volta ao incio e repete todo o processo para o prximo nome da lista, at que todos tenham sido apresentados.

## `while` loops

Diferentemente de `for`, que  um loop definido, `while`  um loop indefinido - ele executa uma ao indefinidamente, **enquanto** (*while*) determinada condio for satisfeita. Ou seja, no cdigo usando `while`, programamos a seguinte instruo: s saia da estrutura de repetio quando a condio no for mais satisfeita.

## Entendendo a sintaxe

````python
while condio:
    # bloco a ser executado
````

### Exemplo
A seguir, um cdigo que acrescenta 1  varivel `count` cada vez que roda o loop, **enquanto** a varivel for menor que 10.

````python
# criar a varivel count com 0 - para que os valores sejam acrescentados  medida que roda o loop
count = 0

# criar o loop
while count<10:          # enquanto count < 10:
    print(count)         # print a varivel
    count = count + 1    # e some 1 ao seu valor atual
````

## Comandos `break` e `continue` - interrompendo um loop

`break`e `continue`so comandos capazes de alterar o fluxo natural do loop, j que s vezes nossa inteno pode ser sair do loop antes que ele chegue ao final.

### `break`
Interrrompe imediatamente o loop mais interno, seja ele `for` ou `while`.
- Quando o loop  interrompido por um `break`, o `else` no  executado.

### `continue`
Interrompe o loop e vai para a prxima iterao (prxima rodada de execuo do loop).

## `if`

````python
# criar a varivel contendo o valor 125
expression_level = 125

if expression_level > 100:                        # se expression_level for maior que 100:
    print("gene is highly expressed")             # printar a frase
````

### Entendendo a sintaxe

- Escrevemos `if` (se) seguida de uma condio e encerramos a linha com `:`
- A seguir, temos um bloco indentado de linhas de cdigo: o **corpo** do `if`
- O corpo s ser executado se a condio for verdadeira.


### Objetivo do `if`

Testar uma propriedade de alguma varivel cujo valor no sabemos quando escrevemos o cdigo. O exemplo acima  apenas ilustrativo, j que sabemos o valor da varivel e ele no ser alterado.

Vamos a outro exemplo:

````python
# definir a lista de nomes de acesso de genes
accs = ['ab56', 'bh84', 'hv76', 'ay93', 'ap97', 'bd72']

# queremos printar apenas os que comeam com "a"
for accession in accs:
    if accession.startswith('a'):
        print(accession)
````

````python
accs = ['ab56', 'bh84', 'hv76', 'ay93', 'ap97', 'bd72']
accs_a = []

# queremos printar apenas os que comeam com "a"
for accession in accs:
    if accession.startswith('a'):
        accs_a.append(accession)
        print(accs_a)
        print(accession)

accs_a
````

### `else`

A clusula `else`  bastante relacionada ao `if`. No exemplo acima, a inteno era apenas gerar uma ao caso a condio fosse verdadeira.

Usamos o `else` quando queremos incluir no programa a opo de fazer uma outra ao, caso a condio no seja verdadeira - assim o programa pode escolher entre duas opes, dependendo da condio.

Para isso, adicionamos o `else` aps o final do corpo do `if`:

````python
expression_level = 125

if expression_level > 130:
    print("gene is highly expressed")
else:
    print("gene is lowly expressed")        # se o expression_level no for maior que 100, printar esta frase.
````

`else` no tem nenhuma condio prpria: ele ser executado quando a condio do `if` for falsa.

No exemplo a seguir, vamos usar `if` e `else` para dividir uma lista de nomes de acesso em dois arquivos diferentes - acessos que comeam com "a" vo para o primeiro arquivo e os outros vo para o segundo arquivo.

````python
# abrir dois arquivos novos para escrever
file1 = open("one.txt", "a") # w = write
file2 = open("two.txt", "a")

# criar a lista de acessos
accs = ['ab56', 'bh84', 'hv76', 'ay93', 'ap97', 'bd72']

# criar o loop
for accession in accs:                         # para cada acesso da lista:
    if accession.startswith('a'):              # se o acesso comear com "a":
        file1.write(accession + "\n")          # escreva em file1 o nome do acesso + nova linha
    else:                                      # caso contrrio (se no comear com "a"):
        file2.write(accession + "\n")          # escreva em file2 o nome do acesso + nova linha

print('Loop concludo.')

# perceba que aqui tambm temos mltiplos nveis de indentao, mas if e else esto no mesmo nvel!
````

## `elif`

At ento viemos analisando apenas uma condio e gerando uma ao caso ela fosse verdadeira. Caso contrrio, outra ao seria gerada usando `else`:

```
if condio:
    o que fazer se a condio for verdadeira
else:
    o que fazer se a condio for falsa
```


O `elif`  usado quando temos mais condies a serem analisadas, da seguinte maneira:

```
if condio_1:
    o que fazer se a condio for verdadeira
elif condio_2:
        o que fazer se a condio_1 for falsa e a condio_2 for verdadeira
else:
    o que fazer se nenhuma das condies acima for verdadeira
```

- Sempre comearemos com `if` e terminaremos com `else` ---> `if` - `elif` - `else`
- Podemos usar quantos `elif` forem necessrios


A seguir, um exemplo com a lista de acesso que j utilizamos antes, mas agora queremos dividir os acessos em trs arquivos em vez de apenas dois: os que comeam com "a" vo no primeiro arquivo, os que comeam com "b" vo no segundo arquivo e os outros vo no terceiro.

````python
# abrir os arquivos novos para escrever:
file1 = open("one.txt", "w")
file2 = open("two.txt", "w")
file3 = open("three.txt", "w")

# criar a lista com os acessos
accs = ['ab56', 'bh84', 'hv76', 'ay93', 'ap97', 'bd72']

# criar o loop
for accession in accs:
    if accession.startswith('a'):          # se o acesso comear com "a":
        file1.write(accession + "\n")      # escreva o nome do acesso em file1
    elif accession.startswith('b'):        # se o acesso no comear com "a" e comear com "b":
        file2.write(accession + "\n")      # escreva o nome do acesso em file2
    else:                                  # caso contrrio (se no comear com "a" nem "b"):
        file3.write(accession + "\n")      # escreva o nome do acesso em file3
````

- Observe que o cdigo continua tendo apenas dois nveis de indentao.
- Podemos acrescentar quantos `elif` quisermos e continuaremos tendo apenas dois nveis.

````python
file4 = open("four.txt", "w")
file5 = open("five.txt", "w")
file6 = open("six.txt", "w")

for accession in accs:
    if accession.startswith('a'):
        file1.write(accession + "\n")
    elif accession.startswith('b'):
        file2.write(accession + "\n")
    elif accession.startswith('c'):
        file3.write(accession + "\n")
    elif accession.startswith('d'):
        file4.write(accession + "\n")
    elif accession.startswith('e'):
        file5.write(accession + "\n")
    # else:
    #     file6.write(accession + "\n")
````

- A estrutura `if` - `elif` - `else`  muito til quando queremos usar vrias opes mutuamente excludentes, ou seja, apenas um ramo pode ser verdadeiro para cada acesso - uma mesma string no pode comear com "a" e "b" ao mesmo tempo.

- Se quisermos escrever um cdigo onde os ramos no sejam mutuamente excludentes (mais de um pode ser verdadeiro), simplesmente acrescentamos vrios `if`:

````python
for accession in accs:
    if accession.startswith('a'):          # se o acesso comear com "a"
        file1.write(accession + "\n")
    if accession.endswith('z'):            # se o acesso terminar com "z"
        file2.write(accession + "\n")
    if len(accession) == 4:                # se o acesso tiver tamanho = 4
        file3.write(accession + "\n")
    if accession.count('j') > 5:           # se o acesso tiver quantidade de "j" maior que 5
        file4.write(accession + "\n")
````

## Construindo condies complexas com `and`, `or` e `not` - operadores booleanos

So usados quando queremos expressar uma condio composta por mais de uma parte.

### `and`
Por exemplo, se quisermos percorrer a lista de acessos e printar os que comeam com "a" e terminam com "3".



````python
# criar a lista de acessos
accs = ['ab56', 'bh84', 'hv76', 'ay93', 'ap97', 'bd72']

# criar o loop
for accession in accs:                                               # para cada acesso da lista:
    if accession.startswith('a') and accession.endswith('3'):        # se o acesso comear com "a" E terminar com "3":
        print(accession)                                             # printar o acesso
````

### `or`
Para produzir uma condio composta por at duas condies simples, que ser verdadeira se uma das duas condies simples forem verdadeiras:

````python
accs = ['ab56', 'bh84', 'hv76', 'ay93', 'ap97', 'bd72']
accs_a_b = []
for accession in accs:
    if accession.startswith('a') or accession.startswith('b'):
        # se o acesso comear com "a" OU comear com "b":
        accs_a_b.append(accession)
        print(accession)

accs_a_b                                             # printar o acesso
````

````python
# criar a lista de acessos
accs = ['ab56', 'bh84', 'hv76', 'ay93', 'ap97', 'bd72']

# criar o loop
for accession in accs:                                               # para cada acesso da lista:
    if accession.startswith('a') or accession.endswith('3'):        # se o acesso comear com "a" E terminar com "3":
        print(accession)
````

````python
# criar uma condio composta que printa acessos que comecem com "a" OU com "b" E terminem com 4:
for acc in accs:
    if (acc.startswith('a') or acc.startswith('b')) and acc.endswith('4'):
        print(acc)
````

Nesse caso, inclumos parnteses para evitar ambiguidades. Se a condio for composta por trs condies simples: X, Y e Z, temos duas opes de condies compostas que significam coisas diferentes:

(X or Y) and Z

no  o mesmo que

X and (Y or Z)

Outra opo seria negar algum tipo de condio usando `not`. O exemplo a seguir vai printar os acessos que comeam com "a" e no terminam com 6.

````python
for acc in accs:
    if acc.startswith('a') and not acc.endswith('6'):
        print(acc)
````
