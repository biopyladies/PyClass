# **Working with files**

Os dados com os quais nós iremos trabalhar são armazenados em arquivos, portanto, se vamos escrever programas úteis, precisamos de uma maneira de obter os dados dos arquivos e colocá-los em nossos programas (e vice-versa).

Escrever uma sequência de DNA diretamente em um programa toda vez que queremos usá-lo não é uma abordagem muito boa a ser adotada. Além disso, parece deselegante misturar os dados com os quais queremos trabalhar com o código que os manipula. Nesta aula, iremos discutir algumas formas de utilização desses arquivos.

### **O que contém os arquivos?**

Nos arquivos que trabalhamos podemos ter vários tipos de dados, mas os principais são os dados de seqüência de DNA e proteína, que podem ser armazenados em uma variedade de formatos. Há muitos outros tipos de dados - leituras de sequenciamento, pontuações de qualidade, SNPs, árvores filogenéticas, mapas de leitura, dados de amostra geográfica, matrizes de distância genética - que podemos acessar de dentro de nossos programas Python.


* Quando se trata de usar o Python no mundo real, geralmente queremos que o **Python aceite ou forneça dados** de outro programa. Freqüentemente, a maneira mais fácil de fazer isso é fazer com que o Python **leia ou grave arquivos em um formato** que o outro programa já entenda.

### **Arquivos de texto**

Na programação, quando falamos de arquivos de texto, estamos falando de um arquivo que contém caracteres e linhas - algo que você pode abrir e visualizar em um editor de texto, independentemente de você realmente entender o arquivo ou não.

Exemplos de arquivos de texto:


*   Arquivos FASTA de sequências de DNA ou proteínas
*   Arquivos contendo saída de programas de linha de comando (por exemplo, BLAST)
*   Arquivos FASTQ contendo leituras de sequenciamento de DNA
*   Arquivos HTML
*   Arquivos em Python



### **Arquivos binários**

São a maioria dos arquivos que você encontra no dia-a-dia, eles são compostos por bytes.

Exemplos de arquivos binários:

* Arquivos de imagem (JPEGs e PNGs)
* Arquivos de áudio
* Arquivos de vídeo
* Arquivos compactados (por exemplo, arquivos ZIP)



## **Criar, abrir e ler arquivos**

Para utilizarmos arquivos externos, devemos passar o nome do arquivo e sua extensão, além do modo que queremos utilizar o arquivo.

Para abrir e ler um arquivo, podemos utilizar a **função open() e o método read()**.

A função **open()** irá abrir o arquivo que passarmos como parâmetro com um determinado **modo de uso** (que também será passado como parâmetro).

Há diversos modos de uso, como podemos ver na imagem abaixo:

image.png

Depois de ler o conteúdo do arquivo em uma variável, podemos tratá-los como qualquer outra string - por exemplo, podemos imprimi-los:

```python
arquivo = open("arquivo_novo.txt", "a") # Aqui estamos criando um arquivo em branco, onde o conteúdo será acrescentado
arquivo.write("Aula 6 - Pyladies") # Conteúdo a ser escrito

# Lembrar de fechar o arquivo!!!...
arquivo.close()
```

```python
# Abrindo o arquivo...
meu_arquivo = open("arquivo_novo.txt", "w") # Aqui substituimos o conteúdo do arquivo
meu_arquivo.write("Aula 6 - Pyladies Bioinfo 2023/2")

# Lembrar de fechar o arquivo!!!...
meu_arquivo.close()
```

## **Gravar e fechar arquivos**

Após abrir um arquivo, é possível alterar o conteúdo e após essa alteração, devemos gravar e fechar o arquivo. Para isso, utilizamos o **método close()**.

Devemos chamar close() depois que terminarmos de ler ou gravar em um arquivo, pois evita alguns tipos de bugs que podem ser difíceis de rastrear --> Boa Prática de Python (Pythonic).

close () é um método incomum, pois não aceita argumentos (por isso é chamado com um par vazio de parênteses) e não retorna nenhum valor útil:

A segunda maneira de fechar um arquivo é usando a **instrução with**.

O comando with automaticamente se encarrega de fechar o arquivo assim que ele sai do bloco with, mesmo em casos de erro.

É recomendado preferir o uso da instrução with, pois permite um **código mais limpo** e facilita o **tratamento de erros inesperados**.

Existe um segundo argumento que é uma string que contém vários caracteres para representar como você deseja abrir o arquivo. O padrão mais comum é **r**, que representa a abertura do arquivo no modo somente leitura como um arquivo de texto:

```python
# Abre o arquivo usando o método with
with open('/content/drive/MyDrive/2 - Grupos de Estudos/2023 1/1. Python Basico/Trabalhando com arquivos/genoma.txt', 'r') as file:
    # Realiza operações no arquivo
    for line in file:
        # Exemplo: imprimir cada linha do arquivo
        print(line)
        # se a linha começa com determinado conjunto de caracteres
        # retornar posições específicas
```

Podemos verificar se o arquivo foi fechado automaticamente:

```python
# Testando se o arquivo foi fechado...
file.closed
```

## **Manipulando arquivos no terminal do Linux**

Conforme trabalhamos com Bioinformática, sentimos a necessidade de melhorar nosso conhecimento sobre sistemas operacionais e como manipular nossos arquivos neles.

Quando falamos em terminal, logo já vem uma imagem de alguma distribuição Linux na cabeça, porém podemos utilizar o terminal no Mac também, alguns comandos são idênticos entre esses sistemas.

No Windows, também temos a possibilidade de utilizar o terminal, nele é chamado de cmd, ou prompt de comando, porém no Linux e no Mac essa prática é mais comum. Por isso, os comando que veremos pelo terminal poderão ser usados no Linux e no Mac, mas não são compatíveis com o Windows.

Não vamos nos aprofundar sobre a linguagem Bash, utilizada no terminal Linux, mas vamos passar alguns comandos úteis, caso precise trabalhar nele.

* Quando abrimos o terminal, na maioria das distribuições Linux, vamos estar na **pasta home** do usuário que estamos utilizando.

* Podemos falar para o terminal mostrar para a gente o diretório atual, ou seja, imprimir o diretório que estamos trabalhando (pwd, print working directory): `pwd`

* Podemos falar para o terminal listar (ls, list) o seu conteúdo: `ls`

* Se queremos mudar de diretório, basta colocar essa informação em inglês (cd, change directory): `cd` `+ nome do diretório de interesse` ou se quisermos sair desse diretório: `cd ..`

* Um comando que é utilizado para criar arquivos no sistema é o `touch`, com ele conseguimos, entre outras coisas, criar arquivos no sistema, exemplo: `estudos.py`

* Para criar um diretório (mkdir, make directory) onde podemos colocar nossos arquivos de estudos: `mkdir Pyladies`

* Nós podemos mover (`mv`) o nosso arquivo `estudos.py` para dentro do diretório `Pyladies`, assim não precisamos criar outro arquivo

* Podemos remover um arquivo (rm, remove) um arquivo: `rm coletando_dados.py` e também diretórios: `rm Pyladies/`. Quando queremos excluir o diretório COM todos os arquivos contidos nele, inserimos o -R de (recursive): `rm -R Pyladies/`
