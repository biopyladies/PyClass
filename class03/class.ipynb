#  **Listas**





> <font color='orangered'>**Uma lista  uma estrutura de dados indexados e armazenados em sequncia, onde cada elemento possui uma posio que  identificada por um ndice.**</font>


- O primeiro elemento  esquerda da lista  o ndice **`0`** e em cada elemento da sequncia o valor do ndice  incrementado **`(+1)`**.

- Ainda  possvel realizar a indexao dos itens por valores negativos, nesse caso a indexao ocorre da direita para esquerda com decremento **`(-1)`**.

image.png

````python
reino = ['animal','planta','fungi','bactria','protista']

print(reino[1])

print(reino[-1])
````

><font color='darkorange'>**Para obter mais de um elemento,  necessrio informar uma posio inicial e uma final, separadas por dois pontos, especificando, assim, o intervalo dos elementos.**</font>

  muito importante lembrar que **os nmeros so inclusivos no fim e excludentes no final**, dessa forma, imprimindo do elemento um **`[1]`** at o quatro **`[4]`**, o resultado sero os elementos de ndice **`[1]`**, **`[2]`** e **`[3]`**.

````python
reino = ['animal', 'planta', 'fungi', 'bactria', 'protista']

print(reino[1:5])
````

### **Adicionar itens  uma lista**

- Para adicionar qualquer elemento a ltima posio de sua lista usa-se o **`append()`**

**Sintaxe**

```python
list.append(elmnt)
```

````python
# Criar a lista
generos_fungos_clavarioides = ['Clavaria', 'Clavulina']

print(generos_fungos_clavarioides)
````

````python
# Adicionar um novo item
generos_fungos_clavarioides.append('Ramaria')

print(generos_fungos_clavarioides) #impresso da lista completa
````

- Para adicionar qualquer elemento em uma posio especfica na lista, usa-se a funo **`insert()`**.

**Sintaxe**

```python
list.insert(pos, elmnt)
```

````python
# Adicionar elemento em uma posio especfica
generos_fungos_clavarioides.insert(2,'Clavulinopsis')
print(generos_fungos_clavarioides)
````

- Para acrescentar elementos de outra lista  lista atual, use o mtodo **`extend()`**.

**Sintaxe**

```python
list.extend(iterable)
```

````python
print(generos_fungos_clavarioides)
````

````python
pterulaceae = ['Pterula', 'Pterulicium']

generos_fungos_clavarioides.extend(pterulaceae)

print(generos_fungos_clavarioides)
````

- Para **alterar** um elemento j existente, fazemos o seguinte processo:

````python
#substituindo o epiteto especfico da orca pelo seu nome popular
generos_fungos_clavarioides[3] = 'Ramariopsis'
print(generos_fungos_clavarioides)
````

### **Remover itens da lista**

- Para remover um elemento existente, podemos usar o mtodo **`remove()`.**

**Sintaxe**

```python
list.remove(elmnt)
```

````python
generos_fungos_clavarioides.remove('Pterula')
print(generos_fungos_clavarioides)
````

 O mtodo **`remove()`** remove a *primeira ocorrncia* do elemento com o valor especificado.

````python
generos_fungos_clavarioides = ['Clavaria', 'Ramaria', 'Clavulina', 'Clavicorona', 'Clavulinopsis', 'Artomyces', 'Clavulina']

generos_fungos_clavarioides.remove('Clavulina')

print(generos_fungos_clavarioides)
````

- Para remover um item de uma posio especfica, podemos usar o mtodo **`pop()`**.

**Sintaxe**

```python
list.pop(pos)
```

````python
generos_fungos_clavarioides.pop(5)
print(generos_fungos_clavarioides)
````

 Se voc no especificar o ndice, o mtodo **`pop()`** remover o ltimo item.

````python
generos_fungos_clavarioides.pop()
print(generos_fungos_clavarioides)
````

- Caso voc precise excluir um intervalo de elementos, basta utilizar a instruo **`del()`**:






````python
generos_fungos_clavarioides = ['Clavaria', 'Ramaria', 'Clavicorona', 'Clavulinopsis', 'Artomyces', 'Clavulina']

# Excluir os elementos do ndice 2 ao ndice 5 (no incluindo o ndice 5)
del generos_fungos_clavarioides[2:5]

print(generos_fungos_clavarioides)
````

  possvel, tambm, esvaziar completamente a lista com o mtodo **`clear()`**.

````python
generos_fungos_clavarioides.clear()
print(generos_fungos_clavarioides)
````

## **Combinar duas listas**

- Existem vrias maneiras de unir ou concatenar duas ou mais listas em Python.

- Uma das maneiras mais fceis  usar o operador **`+`**.


````python
clavariaceae = ['Clavaria', 'Clavulinopsis']
pterulaceae = ['Pterula', 'Pterulicium']

print(clavariaceae)
print(pterulaceae)

agaricales = clavariaceae + pterulaceae

print(agaricales)
````

#  **Tuplas**

><font color='orangered'>As **tuplas** so uma coleo ordenada e imutvel de itens, similar a uma lista, mas seus elementos no podem ser alterados, adicionados ou removidos aps a criao.</font>

## **Como declarar uma tupla?**
 Tuplas so escritas entre **(parnteses)**.

  ```
  procariontes = ('bacteria', 'archaea')
  ```

 *Observao:  O uso da vrgula  exigido, pois se ela tiver apenas um elemento,  preciso colocar uma vrgula depois dele para que o Python entenda que se trata de uma tupla. No caso de uma tupla vazia, a representao segue sendo feita com parnteses sem nenhum elemento.*

 *As tuplas so imutveis, o que significa que no podemos **alterar**, **adicionar** ou **remover** itens aps a criao da tupla.*


````python
eucariontes = ('fungos', 'animais', 'plantas', 'protistas')
print(eucariontes)
````

````python
eucariontes[0]
````

````python
eucariontes[0:4]
````

````python
eucariontes[-1]
````

## **Quando utilizar tuplas?** 

Principais cenrios onde  uma boa ideia usar tuplas:

  * **Dados Constantes e Fixos**: Para informaes que no devem ser alteradas, como datas, coordenadas geogrficas (latitude, longitude), ou configuraes de um sistema.
  * **Retorno de Mltiplos Valores de Funes**: Funes podem retornar uma tupla para entregar vrios resultados de uma vez, como min e max de uma lista, ou o ndice e o valor com enumerate.
  * **Chaves de Dicionrio**: Como so imutveis, tuplas podem ser usadas como chaves em dicionrios (ao contrrio das listas), permitindo agrupar mltiplos itens para uma chave nica.
  * **Estruturas de Dados Heterogneos**: Quando voc tem uma sequncia de itens de tipos diferentes que formam uma "coisa" nica, como um registro de usurio (nome, idade, email).

# **** **Sets**

> <font color='orangered'> Em Python, um set (ou conjunto)  estrutura de dados que armazena uma coleo de **elementos nicos, no ordenados e no indexados**,no ordenada, mutvel, de elementos nicos (sem duplicatas), ideal para realizar operaes matemticas de conjuntos (unio, interseo, diferena) de forma eficiente.</font>








## **Como declarar um conjunto?**</font>

Os conjuntos so escritos entre **{chaves}**:



````python
vitaminas = {'A', 'D', 'E', 'K', 'B1', 'B2', 'B3', 'B5', 'B6', 'B7', 'B9', 'B12', 'C'}
print(vitaminas)
````

````python
# Duplicatas no so permitidas nesta estrutura:

vitaminas = {'A', 'A', 'D', 'E', 'K', 'B1', 'B2', 'B3', 'B5', 'B6', 'B7', 'B9', 'B12', 'C'}
print(vitaminas)
````

````python
vitaminas[0]
````

      <font color='darkorange'>  **IMPORTANTE! Como da para ver no print, sero omitidas ambas 'cpias' duplicadas.**</font>

## **Mutabilidade**

- Os conjuntos so mutveis, o que significa que voc pode adicionar ou remover elementos de um conjunto aps cri-lo.
- No entanto, os prprios elementos devem ser imutveis.
- Isso significa que voc pode adicionar inteiros, strings e tuplas a um conjunto, mas no pode adicionar listas ou outros conjuntos, pois eles so mutveis.

````python
# @title **Adio de elementos**
my_set = {1, 2, 3}

# Adicionando um elemento usando o mtodo add()
my_set.add(4)
print(my_set)
````

````python
# Adicionando elementos usando a notao de atribuio
my_set |= {5, 6}
print(my_set)
````

````python
# @title **Remoo de elementos**

my_set = {1, 2, 3, 4, 5}

# Removendo um elemento usando o mtodo remove()
my_set.remove(3)
print(my_set)  # Sada: {1, 2, 4, 5}

# Removendo um elemento usando o mtodo discard()
my_set.discard(5)
print(my_set)  # Sada: {1, 2, 4}

# Removendo e retornando um elemento usando o mtodo pop()
popped_element = my_set.pop()
print(my_set)          # Sada: {2, 4}
print(popped_element)  # Sada: 1
````

````python
# @title **Limpeza do conjunto**
my_set = {1, 2, 3, 4, 5}

# Limpando o conjunto
my_set.clear()
print(my_set)
````

````python
# Conjunto com elementos mutveis (lista)
my_set = {[1, 2], 3, 4}
````

## **Quando utilizar sets?** 

-  recomendvel a utilizao desse tipo de coleo quando queremos trabalhar com elementos nicos.
- Tambm podemos utilizar os sets quando existe a necessidade de realizar operaes matemticas com elementos de diferentes **`sets`**. Lembre-se que pelo fato de um **`set`** ser no-ordenado, os elementos podem estar em uma ordem diferente a cada execuo do cdigo.
Remover Duplicatas: Para limpar uma lista e obter apenas elementos nicos (ex: de uma lista de IDs de usurios).
- Verificao de Existncia: Para checar se um elemento est presente em uma coleo (como um email cadastrado) mais rpido que em listas.
- Operaes de Conjuntos: Para encontrar itens em comum (interseo), itens diferentes (diferena) ou combinar colees (unio).
- Comparao de Grupos: Comparar dois grupos de itens, como permisses de usurios ou produtos vendidos em diferentes perodos.
- Otimizao de Performance: Para grandes colees onde a velocidade de busca (in)  mais importante que a ordem, pois sets so muito eficientes para isso (tempo constante).

#  **Dicionrios**

> <font color='orangered'>**Dicionrios so estruturas que compreendem um conjunto de pares: `chave` e `valor`.**</font>

## **Como declarar um dicionrio?**

- No que se refere a sintaxe, criar um dicionrio  bem parecido com criar uma lista, mas no lugar de **~colchetes~** usamos **{chaves}**;

- ao invs de elementos, temos itens, **que so separados por vrgulas**;

- dentro de cada item, temos uma **`chave`** e um **`valor`**, que so separados por **dois pontos**.



image.png

````python
taxonomia = {
    "Reino": "Animalia",
    "Filo": "Chordata",
    "Classe": "Mammalia",
    "Ordem": "Carnivora",
    "Famlia": "Felidae",
    "Gnero": "Felis",
    "Espcie": "Felis catus"
}
````


- Para acessar um dado armazenado no dicionrio, basta imprimir a **`chave`** desejada da seguinte forma:

````python
# Acessando e imprimindo o valor de uma categoria especfica
print("Reino:", taxonomia["Reino"])
print("Famlia:", taxonomia["Famlia"])
print("Espcie:", taxonomia["Espcie"])
````

````python
print(taxonomia)
````

 **Os itens do dicionrio so ordenados, mutveis e no permitem duplicatas.**

##**Quando utilizar dicionrios?** 

- Devemos utilizar dicionrios quando existe a necessidade de armazenar dados de forma organizada, visto que, com o dicionrio  possvel mapear o valor de um elemento com uma chave nica.

#  **Resumindo**

Cada estrutura de dados tem suas ventagens e desvantagens.  importante conhecher cada uma delas e entender quais so as nossas necesidades na hora de escolher.

| Caracterstica    | Lista                   | Tupla                   | Conjunto (Set)          | Dicionrio                   |
|-------------------|-------------------------|-------------------------|-------------------------|------------------------------|
| Mutabilidade      | Mutvel                 | Imutvel                | Mutvel                 | Mutvel                      |
| Sintaxe           | `[elemento1, elemento2, ...]` | `(elemento1, elemento2, ...)` | `{elemento1, elemento2, ...}` | `{chave1: valor1, chave2: valor2, ...}` |
| Ordenao         | Ordenada                | Ordenada                | No ordenado            | **Ordenado (Python 3.7+)** / No ordenado (Python 3.6 e anteriores) |
| Elementos Duplicados | Permite               | Permite               | No permite             | No permite                     |
| Acesso aos elementos | Por ndice           | Por ndice           | No suporta ndice      | Por chave                   |
| Sintaxe de iterao | `for elemento in lista:` | `for elemento in tupla:` | `for elemento in conjunto:` | `for chave, valor in dicionrio.items():` |
| Representao vazia | `[]`                    | `()`                    | `set()`                  | `{}`                         |
