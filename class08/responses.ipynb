2. Crie uma **função** que calcule o pH de acordo com a fórmula abaixo.


  Crie uma exceção para lidar com o caso de `H+` ser igual a zero ou negativo (por possíveis erros na manipulação dos números).

  Para saber mais sobre pH e H+: [link](https://pt.khanacademy.org/science/chemistry/acids-and-bases-topic/acids-and-bases/a/ph-poh-and-the-ph-scale)

```python
import math
def calcular_pH(h):
    try:
        pH = -1 * math.log10(h)
        return pH
    except ValueError:
        print("H+ não pode ser zero ou negativo.")

volume = 100
h = 0.01

print(f"O pH é: {calcular_pH(h)}")
```

3. Crie uma função que faça a soma o peso total de uma proteína de acordo com o peso de cada aminoácido. Esta função deve criar um except para caso a proteína contenha algum caracter ("aminoácido") que não é chave do dicionário e outro except para caso algum valor do dicionário não esteja numérico


```python
#exemplo 1
def calcular_massa_proteina(proteina, aminos_massas):
    try:
        # Calcula o peso total da proteína
        peso_total = 0
        for amino_acid in proteina:
          peso_total += aminos_massas[amino_acid]

        return peso_total

    except KeyError:
        print(f"Erro: {amino_acid} é um aminoácido inválido ."
        "Retire-o da sequência e tente novamente")

    except TypeError:
        print(f"Erro: {amino_acid} está com valor não numérico"
        " no banco de dados (dicionário).")


aminos_massas = {
      'A': 71.04, 'C': 103.01, 'D': '115.03', 'E': 129.04, 'F': 147.07,
      'G': 57.02, 'H': 137.06, 'I': 113.08, 'K': 128.09, 'L': 113.08,
      'M': 131.04, 'N': 114.04, 'P': 97.05, 'Q': 128.06, 'R': 156.10,
      'S': 87.03, 'T': 101.05, 'V': 99.07, 'W': 186.08, 'Y': 163.06
}

proteina="AAAAA-"
peso_total = calcular_massa_proteina(proteina, aminos_massas)
if peso_total:
  print(peso_total)
```

⭐ **DESAFIO:** Implemente uma função em Python que calcule o **modelo exponencial** de declínio da população.

O modelo leva em conta apenas a taxa de mortalidade, representada pelo parâmetro `rate`. A função deve receber como parâmetros: `initial_population`, `rate`, `years`.

A função deve retornar a população final.

```python
class SequenciaCurta(Exception):
    pass

def ler_sequencias_fasta(nome_arquivo):
  try:
    with open(nome_arquivo, 'r') as arquivo:
      sequencias = {}
      id_sequencia = ""
      for linha in arquivo:
        print(linha)
        if linha.startswith(">"):
            id_sequencia = linha[1:].strip()
            sequencias[id_sequencia] = ""
        else:
            sequencia = linha.strip()
            if len(sequencia) < 20:
              raise SequenciaCurta(f"A sequência '{sequencia}' contém menos de 20 caracteres)")
              sequencias[id_sequencia] += sequencia
      return sequencias
  except FileNotFoundError:
        print(f"O arquivo '{nome_arquivo}' não foi encontrado.")
        return None


nome_arquivo = input()
sequencias = ler_sequencias_fasta(nome_arquivo)

if sequencias:
    print(sequencias)
```
